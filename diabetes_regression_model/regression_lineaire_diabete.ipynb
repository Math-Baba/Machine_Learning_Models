**Objectif :** Mettre en œuvre une régression linéaire et comprendre l'impact du choix des variables (features) sur la performance du modèle. Nous évaluerons les modèles à l'aide de l'**Erreur Quadratique Moyenne (MSE)** et du **Coefficient de détermination ($R^2$)**.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Charger le dataset
diabetes = load_diabetes()
X_full = diabetes.data
y = diabetes.target

print(f"Forme des données (X) : {X_full.shape}")
print(f"Forme de la cible (y) : {y.shape}")
print(f"Nom des features : {diabetes.feature_names}")
```

```python
# 1. Isoler la feature 'sex'
X_sex = X_full[:, np.newaxis, 1]

# 2. Diviser les données
X_train_sex, X_test_sex, y_train, y_test = train_test_split(X_sex, y, test_size=0.2, random_state=42)

# 3. Créer le modèle
model_sex = LinearRegression()

# 4. Entraîner le modèle
model_sex.fit(X_train_sex, y_train)


# 5. Prédire
predictions = model_sex.predict(X_test_sex)

# 6. Évaluer
print("--- Évaluation (Feature: 'sex') ---")
mse_sex = mean_squared_error(y_test, predictions)
r2_sex = r2_score(y_test, predictions)
print("MSE :", mse_sex)
print("R2 score :", r2_sex)
print("Prédictions :", predictions)
```

```python
# 1. Isoler la feature 'bmi'
X_bmi = X_full[:, np.newaxis, 2]

# 2. Diviser les données
X_train_bmi, X_test_bmi, y_train, y_test = train_test_split(X_bmi, y, test_size=0.2, random_state=42)

# 3. Entraîner un nouveau modèle
model_bmi = LinearRegression()

model_bmi.fit(X_train_bmi, y_train)


# 4. Évaluer
predictions = model_bmi.predict(X_test_bmi)

print("--- Évaluation (Feature: 'bmi') ---")
mse_bmi = mean_squared_error(y_test, predictions)
r2_bmi = r2_score(y_test, predictions)
print("MSE: ", mse_bmi)
print("R2 score: ", r2_bmi)
print("prédictions : ", predictions)
```

```python
# 6. Visualisation
plt.figure(figsize=(8,6))
plt.scatter(X_test_bmi, y_test, color='blue', label='Valeurs réelles')
plt.plot(X_test_bmi, predictions, color='red', linewidth=2, label='Prédictions')
plt.title("Comparaison entre les Prédictions et la Réalité pour 'bmi'")
plt.xlabel("bmi (feature)")
plt.ylabel("Target (y)")
plt.legend()
plt.show()
```

```python
# 1. Diviser les données
X_train, X_test, y_train, y_test = train_test_split(X_full, y, test_size=0.2, random_state=42)

# 2. Entraîner un nouveau modèle
model_global = LinearRegression()

model_global.fit(X_train, y_train)

# 3-4. Évaluer
predictions = model_global.predict(X_test)

print("--- Évaluation (Feature: 'bmi') ---")
mse = mean_squared_error(y_test, predictions)
r2 = r2_score(y_test, predictions)
print("MSE: ", mse)
print("R2 score: ", r2)
print("prédictions : ", predictions)
```

image.png

Dans le cas où :     
* R^2 = 1 donc on a une prédiction parfaite
* R^2 = 0 donc on a un modèle qui ne prédit pas mieux que la moyenne

# Conclusion
Dans la première partie, nous avons créer un modèle de régression linéaire basé sur une seule feature, le sexe de la personne. Nous avons obtenu les résultats suivants :

**MSE :** 5345.981144775686

**score R2 :** -0.009027562540554923

Avec un score R2 négatif, on peut en conclure que le modèle basé uniquement sur le sexe ne peut expliquer la progression de la maladie du diabète. Avec un MSE aussi élévé, on peut donc dire que les valeurs prédites sont bien éloignées des valeurs réelles. En effet, le sexe n'est donc pas un facteur déterminant de la progression de la maladie.

Concernant la deuxième partie, nous avons créer un nouveau modèle basé sur une seule feature toujours mais cette fois ci sur le bmi ou IMC. Nous avons obtenu les résultats suivants:

**MSE:**  4061.8259284949268

**R2 score:**  0.23335039815872138

Avec un score R2 positif, le deuxième modèle permet mieux d'évaluer la progression de la maladie que le premier. Par ailleurs, le MSE est toujours élévé mais moins que celui du premier modèle. Par conséquent, avec le deuxième modèle, on se rapproche un peu plus des valeurs réelles. Cela indique que l'IMC est un facteur déterminant de progression du diabète.

Pour le dernier modèle, nous avons créer un modèle global utilisant toutes les features. Nous avons obtenu les résultats suivants:

**MSE:**  2900.193628493482

**R2 score:**  0.4526027629719195

Ici, on a le meilleur score de R2 des trois modèle se rapprochant de 1. Cela signifie qu'en combinant toutes les caractéristiques, on peut mieux prédire la progression de la maladie plutôt que de n'utiliser qu'une caractéristique. De plus, on a un MSE en dessous de 3000, ce qui est beaucoup moins élevé que les deux autres modèles. Par conséquent, le modèle global est le meilleur pour prédire la progression de la maladie.
